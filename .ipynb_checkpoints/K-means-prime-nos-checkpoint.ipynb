```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
%matplotlib inline
from sklearn.cluster import KMeans

def isPrime(n) : 
  
    # Corner cases 
    if (n <= 1) : 
        return False
    if (n <= 3) : 
        return True
  
    if (n % 2 == 0 or n % 3 == 0) : 
        return False
  
    i = 5
    while(i * i <= n) : 
        if (n % i == 0 or n % (i + 2) == 0) : 
            return False
        i = i + 6
  
    return True

i=0
n = 1
data = []
for i in range(0,10):
    j=0
    row = []
    while(len(row)<2):
        if (isPrime(n)):
            row.append(n)
        n+=1
    data.append(row)
npa = np.asarray(data)
print(data)
plt.scatter(npa[ : , 0], npa[ :, 1], c = 'b', cmap='viridis')
plt.show()
startpts = np.array([[2,3],[5,7]])
Kmean = KMeans(init=startpts, n_clusters=2, verbose=1)
Kmean.fit(npa)
#y_kmeans = KMeans.predict(self, npa)
print(Kmean.cluster_centers_)
plt.scatter(npa[ : , 0], npa[ : , 1], c=Kmean.labels_.astype(float), cmap='viridis')
plt.scatter(11.6, 14.2,  c='g', marker='s')
plt.scatter(49.0, 53.0,  c='r', marker='s')
plt.show()

startpts = np.array([[2,3],[67,71]])
Kmean = KMeans(init=startpts, n_clusters=2, verbose=1)
Kmean.fit(npa)
#y_kmeans = KMeans.predict(self, npa)
print(Kmean.cluster_centers_)
plt.scatter(npa[ : , 0], npa[ : , 1], c=Kmean.labels_.astype(float), cmap='viridis')
plt.scatter(11.6, 14.2,  c='g', marker='s')
plt.scatter(49.0, 53.0,  c='r', marker='s')
plt.show()
```
